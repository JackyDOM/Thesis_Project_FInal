```python
from paddleocr import PaddleOCR, draw_ocr # main OCR dependencies
from matplotlib import pyplot as plt # plot images
import cv2 #opencv
import os # folder directory navigation
from pprint import pprint
```

2. Instantiate Model and Detect

```python
# Setup model
ocr_model = PaddleOCR(lang='en')
```

```python
img_path = os.path.join('.', 'test.png')
```

```python
print(img_path)
```

```python
# Run the ocr method on the ocr model
result = ocr_model.ocr(img_path)
```

```python
pprint(result)
```

```python
pprint(result[0][0][1][0])
```

```python
# Initialize an empty list to store the extracted text
extracted_text = []

# Loop through the result to get the text parts
for block in result:
    for item in block:
        text = item[1][0]  # Extracting the text (first element of the tuple)
        extracted_text.append(text)

# Print all extracted text
for text in extracted_text:
    print(text)
```

3. Visualize Result

```python
# Extracting detected components
boxes = [res[0] for res in result] # 
texts = [res[1][0] for res in result]
scores = [res[1][1] for res in result]
```

```python
# Specifying font path for draw_ocr method
font_path = os.path.join('PaddleOCR', 'doc', 'fonts', 'latin.ttf')
```

```python
# Import our image - drug 1/2/3
# imports image
img = cv2.imread(img_path) 

# reorders the color channels
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
```

```python
# Initialize variables to store the boxes, texts, and scores
boxes = []
texts = []
scores = []

# To store the full address as one string
full_address = ""

# Loop through your OCR result
for block in result:
    for item in block:
        # Get the bounding box, text, and score
        box = item[0]
        text = item[1][0]
        score = item[1][1]
        
        # Append to the boxes, texts, and scores
        boxes.append(box)
        texts.append(text)
        scores.append(score)
        
        # Concatenate the detected text for the address into one string
        # Add a space between words
        full_address += " " + text

# Print the full address
pprint(full_address)

# Visualize the image with annotations
plt.figure(figsize=(15, 15))

# Draw annotations on the image
annotated = draw_ocr(img, boxes, texts, scores, font_path=font_path)

# Show the image with matplotlib
plt.imshow(annotated)
plt.axis('off')  # Hide axes for a cleaner image display
plt.show()
```
